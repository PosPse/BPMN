```python
import json

train_path = "/home/user/baitianrui/bpmn/PAGED/dataset/procedural_graph_extraction/train.json"
test_path = "/home/user/baitianrui/bpmn/PAGED/dataset/procedural_graph_extraction/test.json"
dev_path = "/home/user/baitianrui/bpmn/PAGED/dataset/procedural_graph_extraction/dev.json"

train_data = None
test_data = None
dev_data = None
with open(train_path, "r") as f:
    train_data = json.load(f)

with open(test_path, "r") as f:
    test_data = json.load(f)

with open(dev_path, "r") as f:
    dev_data = json.load(f)
```

```python
dataset = []
for data in [train_data, test_data, dev_data]:
    for item in data:
        temp = {}
        paragraph = item["paragraph"]
        temp["text"] = paragraph
        dataset.append(temp)
```

```python
with open("/home/user/baitianrui/bpmn/PAGED/dataset/procedural_graph_extraction/total.json", "w") as f:
    json.dump(dataset, f)
```
